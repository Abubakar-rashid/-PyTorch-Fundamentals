## What are autograd ?
autograd is PyTorchâ€™s automatic differentiation engine. It automatically tracks operations on tensors and helps compute gradients needed for backpropagation.
## Why autograd?

To understand the need of autograd, we need to understand that it becomes complex to code the derivatives of complex nested functions that are a crucial part of Deep Learning

```python
import torch
```

```python
x = torch.tensor(3.0,requires_grad=True)
# requires_grad=True will tell the pytorch that we are going to calculate gradiant for this
# this param will create a computation graph
```

```python
y = x**2
# y is a function of x
y.backward() #calculates the gradiant of y wrt x
print(x)
print(y)
```

```python
x.grad
## you can only access a leaf node not intermediate node
## for example if x = torch.tensor(3.0,requires_grad=True)
## if y = x**2
## if z = 2y
'''and we calculate z.backward() then we can do y.grad beacuse its and intermediate node but we can do x.grad --> grad of z wrt to x'''
```
